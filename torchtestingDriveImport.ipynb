```
data_transforms = transforms.Compose([
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize([0, 0, 0], [1, 1, 1])
    ])
```

```
image_datasets = torchvision.datasets.ImageFolder(root= "/content/drive/My Drive/Dview wallpapers", transform=data_transforms)
```

```
dataloaders = torch.utils.data.DataLoader(image_datasets, batch_size=1, shuffle=True, num_workers=2)
```

```
dataiter = iter(dataloaders)
imgs = dataiter.next()[0]
print(imgs.shape)
```
